# Exploratory Data Analysis

In this notebook, we will perform exploratory data analysis (EDA) on the dataset. We will load the data, visualize it, and understand its features.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('../data/raw/your_dataset.csv')  # Update with your dataset path

# Display the first few rows of the dataset
data.head()
```

```python
# Summary statistics
data.describe()
```

```python
# Visualize the distribution of a specific feature
plt.figure(figsize=(10, 6))
sns.histplot(data['your_feature'], bins=30, kde=True)  # Update with your feature
plt.title('Distribution of Your Feature')
plt.xlabel('Your Feature')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
