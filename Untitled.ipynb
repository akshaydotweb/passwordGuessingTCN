```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def count_password(file_paths):
    results = {}

    for file_path in file_paths:
        filename = os.path.basename(file_path)

        if file_path.endswith('.txt'):
            try: 
                with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                    passwords = [line.strip() for line in f if line.strip()]
                    count = len(passwords)
            except Exception as e:
                print(f"error processing {filename}: {e}")
                count = 0
        else : 
            print(f"Unsupported file format for {filename}")
            count = 0

        results[filename] = count 

    return results
```

```python
def print_password_counts(results):
    print("\n Password Dataset Analysis")
    print("-" * 50)
    print(f"{'Dataset':<30} | {'Password Count':>15}")
    print("-" * 50)

    total = 0
    for filename, count in results.items():
        print(f"{filename:<30} | {count:>15,}")
        total += count

    print("-" * 50)
    print(f"{'Total':<30} | {total:>15,}")

    return total
```

```python
datasets = [
    "datasets/rockyou.txt",
    "datasets/phpbb.txt",
    "datasets/myspace.txt",
    "datasets/hotmail.txt",
    "datasets/NordVPN.txt",
    "datasets/000webhost.txt",
    "datasets/Ashley-Madison.txt",
    "datasets/singles.org.txt",
    "datasets/10-million-password-list-top-1000000.txt"
]
```

```python
results = count_password(datasets)
```

```python
total = print_password_counts(results)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(results.keys(), results.values())
plt.title(f"Password Dataset Sizes (Total: {total:,} Passwords)")
plt.xlabel("Datasets")
plt.ylabel("Number of Passwords")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def analyze_multiple_password_datasets(file_paths, max_datasets_to_plot=5):
    """
    Analyze and compare password lengths from multiple datasets
    
    Args:
        file_paths: List of paths to password files (.txt or .csv)
        max_datasets_to_plot: Maximum number of datasets to include in plots
    """
    all_results = {}
    all_stats = []
    
    # Process each dataset
    for file_path in file_paths:
        file_name = os.path.basename(file_path)
        print(f"\nProcessing {file_name}...")
        
        try:
            # Load passwords based on file type
            passwords = []
            
            if file_path.endswith('.txt'):
                with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                    passwords = [line.strip() for line in f if line.strip()]
            
            elif file_path.endswith('.csv'):
                df = pd.read_csv(file_path)
                password_col = 'password' if 'password' in df.columns else df.columns[0]
                passwords = df[password_col].dropna().tolist()
            
            else:
                print(f"Skipping {file_name}: Unsupported file format. Use .txt or .csv files.")
                continue
                
            # Calculate length of each password
            lengths = [len(pwd) for pwd in passwords]
            
            # Count frequency of each length
            length_counts = Counter(lengths)
            
            # Sort by length
            sorted_lengths = sorted(length_counts.items())
            
            # Create dataframe for analysis
            length_df = pd.DataFrame(sorted_lengths, columns=['Length', 'Frequency'])
            length_df['Percentage'] = 100 * length_df['Frequency'] / length_df['Frequency'].sum()
            
            # Store results
            all_results[file_name] = length_df
            
            # Store statistics
            stats = {
                'Dataset': file_name,
                'Total Passwords': len(passwords),
                'Average Length': sum(lengths)/len(lengths),
                'Min Length': min(lengths),
                'Max Length': max(lengths),
                'Most Common Length': length_df.loc[length_df['Frequency'].idxmax(), 'Length']
            }
            all_stats.append(stats)
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
    
    # Create statistics summary
    stats_df = pd.DataFrame(all_stats)
    print("\n=== Password Dataset Statistics ===")
    print(stats_df.to_string(index=False, float_format=lambda x: f"{x:.2f}"))
    
    # Select top datasets for visualization (to avoid overcrowding)
    top_datasets = sorted(all_results.items(), 
                         key=lambda x: stats_df[stats_df['Dataset'] == x[0]]['Total Passwords'].values[0], 
                         reverse=True)[:max_datasets_to_plot]
    
    # Create comparison visualizations
    plt.figure(figsize=(15, 10))
    
    # 1. Length distribution comparison
    plt.subplot(2, 1, 1)
    for name, df in top_datasets:
        plt.plot(df['Length'], df['Percentage'], marker='.', linestyle='-', label=name)
    
    plt.title('Password Length Distribution Comparison')
    plt.xlabel('Password Length')
    plt.ylabel('Percentage')
    plt.grid(alpha=0.3)
    plt.legend()
    
    # 2. Box plot comparison
    plt.subplot(2, 1, 2)
    box_data = []
    labels = []
    
    for name, df in top_datasets:
        # Reconstruct raw data for box plot (approximate)
        lengths = []
        for length, freq in zip(df['Length'], df['Frequency']):
            lengths.extend([length] * int(freq))
        
        # Sample to keep box plot efficient
        max_samples = 10000
        if len(lengths) > max_samples:
            lengths = np.random.choice(lengths, max_samples, replace=False)
            
        box_data.append(lengths)
        labels.append(name)
    
    plt.boxplot(box_data, labels=labels)
    plt.title('Password Length Distribution (Box Plot)')
    plt.ylabel('Password Length')
    plt.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig("password_length_comparison.png")
    plt.show()
    
    return all_results, stats_df

# Example usage
datasets = [
    'datasets/rockyou.txt'
]

results, stats = analyze_multiple_password_datasets(datasets)
```
