```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def count_password(file_paths):
    results = {}

    for file_path in file_paths:
        filename = os.path.basename(file_path)

        if file_path.endswith('.txt'):
            try: 
                with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                    passwords = [line.strip() for line in f if line.strip()]
                    count = len(passwords)
            except Exception as e:
                print(f"error processing {filename}: {e}")
                count = 0
        else : 
            print(f"Unsupported file format for {filename}")
            count = 0

        results[filename] = count 

    return results
```

```python
def print_password_counts(results):
    print("\n Password Dataset Analysis")
    print("-" * 50)
    print(f"{'Dataset':<30} | {'Password Count':>15}")
    print("-" * 50)

    total = 0
    for filename, count in results.items():
        print(f"{filename:<30} | {count:>15,}")
        total += count

    print("-" * 50)
    print(f"{'Total':<30} | {total:>15,}")

    return total
```

```python
datasets = [
    "datasets/rockyou.txt",
    "datasets/phpbb.txt",
    "datasets/myspace.txt",
    "datasets/hotmail.txt",
    "datasets/NordVPN.txt",
    "datasets/000webhost.txt",
    "datasets/Ashley-Madison.txt",
    "datasets/singles.org.txt",
    "datasets/10-million-password-list-top-1000000.txt"
]
```

```python
results = count_password(datasets)
```

```python
total = print_password_counts(results)
```

```python
from collections import Counter

def analyze_multiple_password_datasets(file_paths, max_datasets_to_plot=5):
    all_results = {}
    all_stats = []
    
    # Process each dataset
    for file_path in file_paths:
        file_name = os.path.basename(file_path)
        print(f"\nProcessing {file_name}...")
        
        try:
            # Load passwords based on file type
            passwords = []
            
            if file_path.endswith('.txt'):
                with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                    passwords = [line.strip() for line in f if line.strip()]
            
            elif file_path.endswith('.csv'):
                df = pd.read_csv(file_path)
                password_col = 'password' if 'password' in df.columns else df.columns[0]
                passwords = df[password_col].dropna().tolist()
            
            else:
                print(f"Skipping {file_name}: Unsupported file format. Use .txt or .csv files.")
                continue
                
            # Calculate length of each password
            lengths = [len(pwd) for pwd in passwords]
            
            # Count frequency of each length
            length_counts = Counter(lengths)
            
            # Sort by length
            sorted_lengths = sorted(length_counts.items())
            
            # Create dataframe for analysis
            length_df = pd.DataFrame(sorted_lengths, columns=['Length', 'Frequency'])
            length_df['Percentage'] = 100 * length_df['Frequency'] / length_df['Frequency'].sum()
            
            # Store results
            all_results[file_name] = length_df
            
            # Store statistics
            stats = {
                'Dataset': file_name,
                'Total Passwords': len(passwords),
                'Average Length': sum(lengths)/len(lengths),
                'Min Length': min(lengths),
                'Max Length': max(lengths),
                'Most Common Length': length_df.loc[length_df['Frequency'].idxmax(), 'Length']
            }
            all_stats.append(stats)
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
    
    # Create statistics summary
    stats_df = pd.DataFrame(all_stats)
    print("\n=== Password Dataset Statistics ===")
    print(stats_df.to_string(index=False, float_format=lambda x: f"{x:.2f}"))
    
    # Select top datasets for visualization (to avoid overcrowding)
    top_datasets = sorted(all_results.items(), 
                         key=lambda x: stats_df[stats_df['Dataset'] == x[0]]['Total Passwords'].values[0], 
                         reverse=True)[:max_datasets_to_plot]
    
    # Create comparison visualizations
    plt.figure(figsize=(15, 10))
    
    # 1. Length distribution comparison
    plt.subplot(2, 1, 1)
    for name, df in top_datasets:
        plt.plot(df['Length'], df['Percentage'], marker='.', linestyle='-', label=name)
    
    plt.title('Password Length Distribution Comparison')
    plt.xlabel('Password Length')
    plt.ylabel('Percentage')
    plt.grid(alpha=0.3)
    plt.legend()
    
    # 2. Box plot comparison
    plt.subplot(2, 1, 2)
    box_data = []
    labels = []
    
    for name, df in top_datasets:
        # Reconstruct raw data for box plot (approximate)
        lengths = []
        for length, freq in zip(df['Length'], df['Frequency']):
            lengths.extend([length] * int(freq))
        
        # Sample to keep box plot efficient
        max_samples = 10000
        if len(lengths) > max_samples:
            lengths = np.random.choice(lengths, max_samples, replace=False)
            
        box_data.append(lengths)
        labels.append(name)
    
    plt.boxplot(box_data, labels=labels)
    plt.title('Password Length Distribution (Box Plot)')
    plt.ylabel('Password Length')
    plt.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig("password_length_comparison.png")
    plt.show()
    
    return all_results, stats_df


results, stats = analyze_multiple_password_datasets(datasets)
```

```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import re

def analyze_password_datasets(file_paths):
    all_results = []
    
    # Process each dataset
    for file_path in file_paths:
        file_name = os.path.basename(file_path)
        print(f"\nAnalyzing {file_name}...")
        
        try:
            # Load passwords based on file type
            passwords = []
            
            if file_path.endswith('.txt'):
                with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                    passwords = [line.strip() for line in f if line.strip()]
            
            elif file_path.endswith('.csv'):
                df = pd.read_csv(file_path)
                password_col = 'password' if 'password' in df.columns else df.columns[0]
                passwords = df[password_col].dropna().tolist()
            
            else:
                print(f"Skipping {file_name}: Unsupported file format")
                continue
            
            # Count total passwords
            total_passwords = len(passwords)
            
            # Calculate length statistics
            lengths = [len(pwd) for pwd in passwords]
            length_counts = Counter(lengths)
            
            avg_length = sum(lengths) / total_passwords if total_passwords > 0 else 0
            min_length = min(lengths) if lengths else 0
            max_length = max(lengths) if lengths else 0
            
            # Classify passwords by type
            only_alpha = 0
            only_numeric = 0
            only_symbols = 0
            alphanumeric = 0
            alphanumeric_with_symbols = 0
            
            for pwd in passwords:
                if re.match(r'^[a-zA-Z]+$', pwd):
                    only_alpha += 1
                elif re.match(r'^[0-9]+$', pwd):
                    only_numeric += 1
                elif re.match(r'^[^a-zA-Z0-9]+$', pwd):
                    only_symbols += 1
                elif re.match(r'^[a-zA-Z0-9]+$', pwd):
                    alphanumeric += 1
                else:
                    alphanumeric_with_symbols += 1
            
            # Store results
            result = {
                'Dataset': file_name,
                'Total Passwords': total_passwords,
                'Average Length': avg_length,
                'Min Length': min_length,
                'Max Length': max_length,
                'Only Alphabets': only_alpha,
                'Only Alphabets %': 100 * only_alpha / total_passwords if total_passwords > 0 else 0,
                'Only Numbers': only_numeric,
                'Only Numbers %': 100 * only_numeric / total_passwords if total_passwords > 0 else 0,
                'Only Symbols': only_symbols,
                'Only Symbols %': 100 * only_symbols / total_passwords if total_passwords > 0 else 0,
                'Alphanumeric': alphanumeric,
                'Alphanumeric %': 100 * alphanumeric / total_passwords if total_passwords > 0 else 0,
                'Alpha+Num+Symbol': alphanumeric_with_symbols,
                'Alpha+Num+Symbol %': 100 * alphanumeric_with_symbols / total_passwords if total_passwords > 0 else 0,
                'Length Distribution': length_counts
            }
            
            all_results.append(result)
            
            # Print stats
            print(f"  Total passwords: {total_passwords:,}")
            print(f"  Average length: {avg_length:.2f} characters")
            print(f"  Length range: {min_length} to {max_length} characters")
            print("  Password categories:")
            print(f"    - Only alphabets: {only_alpha:,} ({result['Only Alphabets %']:.2f}%)")
            print(f"    - Only numbers: {only_numeric:,} ({result['Only Numbers %']:.2f}%)")
            print(f"    - Only symbols: {only_symbols:,} ({result['Only Symbols %']:.2f}%)")
            print(f"    - Alphanumeric: {alphanumeric:,} ({result['Alphanumeric %']:.2f}%)")
            print(f"    - Alpha+Num+Symbol: {alphanumeric_with_symbols:,} ({result['Alpha+Num+Symbol %']:.2f}%)")
            
            # Create visualizations for each dataset
            create_dataset_visualizations(result, file_name)
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
    
    # Create summary table
    summary = pd.DataFrame([{
        'Dataset': r['Dataset'],
        'Total Passwords': r['Total Passwords'],
        'Avg Length': r['Average Length'],
        'Only Alpha %': r['Only Alphabets %'],
        'Only Num %': r['Only Numbers %'],
        'Only Symbol %': r['Only Symbols %'],
        'Alphanum %': r['Alphanumeric %'],
        'Complex %': r['Alpha+Num+Symbol %']
    } for r in all_results])
    
    print("\n=== Password Dataset Summary ===")
    print(summary.to_string(index=False, float_format=lambda x: f"{x:.2f}"))
    
    # Create comparison visualization if multiple datasets
    if len(all_results) > 1:
        create_comparison_visualizations(all_results)
    
    return all_results, summary

def create_dataset_visualizations(result, file_name):
    """Create visualizations for a single dataset"""
    plt.figure(figsize=(15, 10))
    
    # 1. Length distribution
    plt.subplot(2, 2, 1)
    lengths = sorted(result['Length Distribution'].keys())
    frequencies = [result['Length Distribution'][k] for k in lengths]
    plt.bar(lengths, frequencies, color='skyblue')
    plt.title(f'Password Length Distribution: {file_name}')
    plt.xlabel('Password Length')
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.3)
    
    # 2. Password type distribution
    plt.subplot(2, 2, 2)
    categories = ['Only\nAlpha', 'Only\nNumbers', 'Only\nSymbols', 
                 'Alpha+\nNum', 'Complex']
    values = [
        result['Only Alphabets'],
        result['Only Numbers'],
        result['Only Symbols'],
        result['Alphanumeric'],
        result['Alpha+Num+Symbol']
    ]
    plt.bar(categories, values, color=['green', 'orange', 'red', 'purple', 'blue'])
    plt.title(f'Password Type Distribution: {file_name}')
    plt.ylabel('Count')
    plt.grid(axis='y', alpha=0.3)
    
    # 3. Password type pie chart
    plt.subplot(2, 2, 3)
    plt.pie(values, labels=categories, autopct='%1.1f%%',
            startangle=90, colors=['green', 'orange', 'red', 'purple', 'blue'])
    plt.axis('equal')
    plt.title(f'Password Type Percentage: {file_name}')
    
    # 4. Cumulative length distribution
    plt.subplot(2, 2, 4)
    sorted_lengths = sorted(result['Length Distribution'].items())
    x = [item[0] for item in sorted_lengths]
    y = [item[1] for item in sorted_lengths]
    cumulative = np.cumsum(y)
    cumulative_percent = 100 * cumulative / cumulative[-1] if cumulative[-1] > 0 else []
    
    plt.plot(x, cumulative_percent, marker='o', color='green')
    plt.title(f'Cumulative Length Distribution: {file_name}')
    plt.xlabel('Password Length')
    plt.ylabel('Cumulative %')
    plt.grid(alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f"{os.path.splitext(file_name)[0]}_analysis.png")
    plt.close()
    
def create_comparison_visualizations(all_results):
    """Create visualizations comparing multiple datasets"""
    plt.figure(figsize=(15, 10))
    
    # Get dataset names and prepare data
    datasets = [result['Dataset'] for result in all_results]
    
    # 1. Compare complexity distributions
    plt.subplot(2, 2, 1)
    data = np.array([[
        result['Only Alphabets %'],
        result['Only Numbers %'],
        result['Only Symbols %'],
        result['Alphanumeric %'],
        result['Alpha+Num+Symbol %']
    ] for result in all_results])
    
    categories = ['Only Alpha', 'Only Numeric', 'Only Symbols', 
                 'Alphanumeric', 'Complex']
    
    x = np.arange(len(datasets))
    width = 0.15
    multiplier = 0
    
    for i, category in enumerate(categories):
        offset = width * multiplier
        plt.bar(x + offset, data[:, i], width, label=category)
        multiplier += 1
        
    plt.title('Password Complexity Distribution')
    plt.xticks(x + width * 2, datasets, rotation=45, ha='right')
    plt.ylabel('Percentage')
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=5)
    
    # 2. Compare average lengths
    plt.subplot(2, 2, 2)
    avg_lengths = [result['Average Length'] for result in all_results]
    plt.bar(datasets, avg_lengths, color='green')
    plt.title('Average Password Length')
    plt.ylabel('Characters')
    plt.grid(axis='y', alpha=0.3)
    plt.xticks(rotation=45, ha='right')
    
    plt.tight_layout()
    plt.savefig("password_comparison.png")
    plt.close()

results, summary = analyze_password_datasets(datasets)
```

```python
def plot_password_histograms(file_paths, bins=20, figsize=(15, 10), sample_size=None):
    num_datasets = len(file_paths)
    
    # Create subplot grid
    rows = int(np.ceil(num_datasets / 2))
    cols = 2 if num_datasets > 1 else 1
    
    plt.figure(figsize=figsize)
    
    for idx, file_path in enumerate(file_paths):
        try:
            # Load passwords
            file_name = os.path.basename(file_path)
            passwords = []
            
            if file_path.endswith('.txt'):
                with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                    passwords = [line.strip() for line in f if line.strip()]
            elif file_path.endswith('.csv'):
                df = pd.read_csv(file_path)
                password_col = 'password' if 'password' in df.columns else df.columns[0]
                passwords = df[password_col].dropna().tolist()
            else:
                print(f"Skipping {file_name}: Unsupported file format")
                continue
                
            # Sample if needed
            if sample_size and len(passwords) > sample_size:
                passwords = np.random.choice(passwords, sample_size, replace=False)
            
            # Calculate lengths
            lengths = [len(pwd) for pwd in passwords]
            
            # Create histogram
            plt.subplot(rows, cols, idx + 1)
            n, bins, patches = plt.hist(lengths, bins=bins, alpha=0.7, color='skyblue', 
                                        edgecolor='black')
            
            # Add statistics
            avg_len = np.mean(lengths)
            median_len = np.median(lengths)
            
            plt.axvline(avg_len, color='red', linestyle='dashed', linewidth=1, 
                       label=f'Mean: {avg_len:.1f}')
            plt.axvline(median_len, color='green', linestyle='dashed', linewidth=1,
                       label=f'Median: {median_len:.1f}')
            
            # Customize plot
            plt.title(f'Password Length Distribution: {file_name}')
            plt.xlabel('Password Length')
            plt.ylabel('Frequency')
            plt.grid(alpha=0.3)
            plt.legend()
            
            # Add text with statistics
            stats_text = f"Total: {len(passwords):,}\nAvg: {avg_len:.2f}\n" \
                         f"Min: {min(lengths)}\nMax: {max(lengths)}"
            plt.text(0.95, 0.95, stats_text, transform=plt.gca().transAxes, 
                    fontsize=9, verticalalignment='top', horizontalalignment='right',
                    bbox={'boxstyle': 'round', 'facecolor': 'white', 'alpha': 0.8})
            
        except Exception as e:
            plt.subplot(rows, cols, idx + 1)
            plt.text(0.5, 0.5, f"Error processing {file_name}:\n{str(e)}", 
                    ha='center', va='center', transform=plt.gca().transAxes)
    
    plt.tight_layout()
    plt.savefig("password_histograms.png", dpi=300)
    plt.show()

# Simple histogram with default bins
plot_password_histograms(datasets)

# Histogram with custom bins
custom_bins = list(range(0, 31, 2))  # 0-30 in steps of 2
plot_password_histograms(datasets, bins=custom_bins, figsize=(16, 12))

# Histogram with sampling (for very large datasets)
plot_password_histograms(datasets, sample_size=10000)
```
